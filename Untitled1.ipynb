```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_excel(list(uploaded.keys())[0])
print(df.head())
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.fillna(df.mean())
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.dtypes
```

```python
X = df.drop("Class", axis=1)
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

print("Model Trained Successfully!")
```

```python
def predict_fraud(input_list):
    scaled = scaler.transform([input_list])
    pred = model.predict(scaled)[0]
    return "FRAUD" if pred == 1 else "NOT FRAUD"
    print("\nEnter transaction details to predict:")
```

```python
sample = []

for col in X.columns:
    value_str = input(f"Enter value for {col}: ")
    value = float(value_str.replace(',', '.'))
    sample.append(value)

result = predict_fraud(sample)
print("\nPrediction:", result)
```
