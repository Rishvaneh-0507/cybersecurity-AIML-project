```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel("creditcard.xlsx")

plt.figure(figsize=(12, 8))

# ---------- 1. Line Plot ----------
plt.subplot(2, 2, 1)
plt.plot(df['V1'][:500], df['V2'][:500], color='red')   # limit rows for visibility
plt.title("Line Plot (V1 vs V2)")
plt.xlabel("V1")
plt.ylabel("V2")

# ---------- 2. Pie Chart ----------
plt.subplot(2, 2, 2)
pie_values = df['V2'][:5].abs()   # pie chart cannot take negative values
plt.pie(pie_values, labels=df['V1'][:5], autopct="%1.1f%%")
plt.title("Pie Chart (First 5 Values)")

# ---------- 3. Bar Chart ----------
plt.subplot(2, 2, 3)
plt.bar(df['V1'][:10], df['V2'][:10], color='red')
plt.title("Bar Chart (10 Samples)")
plt.xlabel("V1")
plt.ylabel("V2")

# ---------- 4. Scatter Plot ----------
plt.subplot(2, 2, 4)
plt.scatter(df['V1'][:500], df['V2'][:500], color='red')  # limit data for clarity
plt.title("Scatter Plot (V1 vs V2)")
plt.xlabel("V1")
plt.ylabel("V2")

plt.tight_layout()
plt.savefig("dashboard.png",dpi=300,bbox_inches='tight')
plt.show()
```
